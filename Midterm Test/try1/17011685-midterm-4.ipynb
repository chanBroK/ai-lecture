```python
import torch
import sklearn
import random
# 장치 설정
device = "cuda" if torch.cuda.is_available() else "cpu"

# 난수 시드 설정
torch.manual_seed(1)
random.seed(1)
if device == "cuda":
    torch.cuda.manual_seed_all(1)
    
# 데이터 로드
train = pd.read_csv("/kaggle/input/2021-ai-midterm-p4/train.csv")
test = pd.read_csv("/kaggle/input/2021-ai-midterm-p4/test.csv")
submission = pd.read_csv("/kaggle/input/2021-ai-midterm-p4/submit_sample.csv")

#데이터 확인
print(train.head())
print(test.head())
print(submission.head())

# 데이터 전처리
from sklearn.preprocessing import StandardScaler, LabelEncoder,MinMaxScaler

# 난측값 채우기
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
y_train = np.array(train['target'])

train = pd.DataFrame(imputer.fit_transform(train.drop(['target'],axis=1)))
test = pd.DataFrame(imputer.transform(test))

#데이터 확인
print(train.head())
print(test.head())
print(submission.head())

# 실수화 
le = LabelEncoder()
# city 처리
train[2]=le.fit_transform(train[2])
test[2]=le.fit_transform(test[2])
# gender 처리
train[4]=le.fit_transform(train[4])
test[4]=le.fit_transform(test[4])
# relevent_experience  처리
train[5]=le.fit_transform(train[5])
test[5]=le.fit_transform(test[5])
# enrolled_university  처리
train[6]=le.fit_transform(train[6])
test[6]=le.fit_transform(test[6])
# education_level  처리
train[7]=le.fit_transform(train[7])
test[7]=le.fit_transform(test[7])

# major_discipline   처리
train[8]=le.fit_transform(train[8])
test[8]=le.fit_transform(test[8])

# company_type  처리
train[11]=le.fit_transform(train[11])
test[11]=le.fit_transform(test[11])

# 필요없는 필드 삭제
x_train = np.array(train.drop([0,9,10,12],axis=1))
x_test = np.array(test.drop([0,9,10,12],axis=1))


# 데이터 변형(scale)
# scaler = StandardScaler()
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)


# 데이터 텐서에 올리기
x_train = torch.Tensor(x_train).to(device)
x_test = torch.Tensor(x_test).to(device)
y_train = torch.Tensor(y_train).to(device)

# 데이터 셋 
from torch.utils.data import DataLoader,TensorDataset
train_dataset = TensorDataset(x_train,y_train)

# 데이터 모양 확인
print(x_train.shape , y_train.shape)

# Layer 설정
layer1 = torch.nn.Linear(10,20,bias =True).to(device)
layer2 = torch.nn.Linear(20,40,bias =True).to(device)
layer3 = torch.nn.Linear(40,20,bias =True).to(device)
layer4 = torch.nn.Linear(20,10,bias =True).to(device)
layer5 = torch.nn.Linear(10,1,bias =True).to(device)
relu = torch.nn.ReLU().to(device)
dropout = torch.nn.Dropout(0.3).to(device)

# Layer 초기화
torch.nn.init.xavier_normal_(layer1.weight)
torch.nn.init.xavier_normal_(layer2.weight)
torch.nn.init.xavier_normal_(layer3.weight)
torch.nn.init.xavier_normal_(layer4.weight)
torch.nn.init.xavier_normal_(layer5.weight)


# Model 설정
model = torch.nn.Sequential(layer1,relu,dropout,
                            layer2,relu,dropout,
                            layer3,relu,dropout,
                            layer4,relu,dropout,
                            layer5).to(device)

# 학습 파라미터 설정
epochs = 1000
lr = 0.01
batch_size = 500

# loss 함수 이진 분류
# loss = torch.nn.CrossEntropyLoss()
# loss = torch.nn.MSELoss()
loss = torch.nn.BCELoss()

# optim 
optim = torch.optim.Adam(model.parameters(),lr=lr)

#data loader
data_loader = DataLoader(dataset=train_dataset,batch_size=batch_size)

# 학습
model.train()
for epoch in range(epochs):
    optim.zero_grad()
    h = model(x_train)
    cost = loss(torch.sigmoid(h),y_train.unsqueeze(1))
    cost.backward()
    optim.step()
    if epoch % 100 == 0:
        print(epoch , cost.item())
```

```python
# 학습 정확도
with torch.no_grad():
    model.eval()
    predict = model(x_train)
    predict = torch.sigmoid(predict)
    predict = predict >= torch.Tensor([0.5]).to(device)
    predict = predict.int()
    
    acc = predict == y_train
    print(acc)
    print(acc.sum())
    print(acc.sum()/len(predict))
```

```python
# 모델 예측 및 저장
with torch.no_grad():
    model.eval()
    predict = model(x_test)
    predict = torch.sigmoid(predict)
    predict = predict >= torch.Tensor([0.5]).to(device)
    predict = predict.float().cpu().detach().numpy()
    submission['target'] = predict
submission.to_csv("submission.csv",index=False)
```
